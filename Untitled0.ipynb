```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Bar plot
plt.bar(x, y, color='red')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar Plot')
plt.show()

# Box plot
plt.boxplot(y)
plt.show()

# Line plot
plt.plot(x, y)
plt.show()

# Histogram
plt.hist(y)
plt.show()

# Scatter plot
plt.scatter(x, y)
plt.show()

# Area plot
plt.fill_between(x, y)
plt.show()

# Pie chart
plt.pie(y)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("sample_data/college.csv")

# Bar plot
plt.bar(df["college_id"], df["city"],color="green")
plt.xlabel("team_id")
plt.ylabel("g")
plt.title("Team goals")
plt.show()

# Box plot
sns.boxplot(x=df["city"])
plt.show()

# Line plot
plt.plot(df["college_id"], df["city"])
plt.show()

# Histogram
plt.hist(df["city"])
plt.show()

# Scatter plot
plt.scatter(df["college_id"], df["city"])
plt.show()

# Area plot
plt.fill_between(df["college_id"], df["city"])
plt.show()

# Pie chart
plt.pie(df["number"])
```
